```python
# Escriba una funcion que reciba un arbol binario y un valor a encontrar.Retorne cuantas veces est este valor en todo el arbol omitiendo las hojas
import random
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinaryTreeS:
    def __init__(self):
        self.root = None


    def insert(self, value, current):
        #hay raíz?
        if(self.root is None):
            self.root = Node(value)
            return
        if(current.value == value):
            return
        elif(value > current.value):
            if(current.right is None):
                current.right = Node(value)
                return
            else:
                self.insert(value,current.right)
        else:
            if(current.left is None):
                current.left = Node(value)
                return
            else:
                self.insert(value,current.left)


    def pretty_print_tree(self, node, prefix="", is_left=True):
        if not node:
            print("Empty Tree")
            return
        if node.right:
            self.pretty_print_tree(node.right, prefix + ("│   " if is_left else "    "), False)
        print(prefix + ("└── " if is_left else "┌── ") + str(node.value))
        if node.left:
            self.pretty_print_tree(node.left, prefix + ("    " if is_left else "│   "), True)

def search_bt(bt: BinaryTree, current, find, cont = 0, l = []):
    if bt.root == None:
        return
    else:
        if current.right != None:
            curd = current.right
            if curd.value == find and curd.right != None and curd.left != None:
                cont += 1
            l.append(current.right)
        if current.left != None:
            curl = current.left
            if curl.value == find and curl.right != None and curl.left:
                cont += 1
            l.append(current.left)
        if len(l) > 0:
            return search_bt(bt, l.pop(0), find, cont, l)
        else:
            if bt.root.value == find:
                cont += 1
            return cont

bt = BinaryTree()
for i in range(7):
    r = random.randrange(0, 3)
    bt.add(bt.root,r)
bt.pretty_print_tree(bt.root)

print(search_bt(bt, bt.root, 2))
```
